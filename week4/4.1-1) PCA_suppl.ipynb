```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n_data = 100
dim = 50
mean_vec = np.zeros([dim])
Cov_mat = np.random.randn(dim**2).reshape([dim,dim])
Cov_mat = np.matmul(Cov_mat.T, Cov_mat)

X = np.random.multivariate_normal(mean_vec, Cov_mat, n_data)
```

```python
fig = plt.figure(figsize = (10, 8))
plt.plot(X[:,0], X[:,1], 'k.', alpha = 0.3)
plt.axis('equal')
plt.grid(alpha = 0.3)
plt.show()
```

```python
mean_X = np.mean(X, axis=0)

C = np.matmul((X - mean_X).T, (X - mean_X))/n_data
C = (C + C.T)/2
e_val, e_vec = np.linalg.eig(C)
plt.plot(np.sort(e_val))
```

```python
coords = np.matmul(X, e_vec[:,np.argsort(e_val)[[-2,-1]]])
plt.scatter(coords[:,0], coords[:,1])
```

### Kernelize

```python
K = np.matmul((X - mean_X), (X - mean_X).T)
K = (K + K.T)/2
e_val, e_vec = np.linalg.eig(K)
plt.plot(np.sort(e_val)/n_data)
```

```python
coords = np.matmul(K, np.asarray(e_vec[:,np.argsort(np.asarray(e_val, float))[[-2,-1]]], float))
plt.scatter(coords[:,0], coords[:,1])
```
